```python
import pandas as pd 
import sklearn
from sklearn.impute import SimpleImputer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_prime= pd.read_csv("../DSCI 551 Project/Amazon Prime/titles.csv")
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
df_prime['imdb_score']=imputer.fit_transform(df_prime['imdb_score'].to_numpy().reshape(-1,1))
df_prime_fnl=df_prime.round({"imdb_score":1})
```

```python
df_disney= pd.read_csv("../DSCI 551 Project/Disney/titles.csv")
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
df_disney['imdb_score']=imputer.fit_transform(df_disney['imdb_score'].to_numpy().reshape(-1,1))
df_disney_fnl=df_disney.round({"imdb_score":1})
```

```python
df_netflix= pd.read_csv("../DSCI 551 Project/Netflix/titles.csv")
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
df_netflix['imdb_score']=imputer.fit_transform(df_netflix['imdb_score'].to_numpy().reshape(-1,1))
df_netflix_fnl=df_netflix.round({"imdb_score":1})
```

```python
df=pd.concat([df_prime_fnl,df_disney_fnl,df_netflix_fnl])
```

```python
df
```

```python
df_tbl=df['genres'].tolist()
```

```python
df_tbl
```

```python
l_s=[sub.replace('[]','["U/A Movie"]') for sub in df_tbl]
```

```python
l_s
```

```python
df['genres']=l_s
```

```python
x1=pd.DataFrame(df_disney['release_year'].value_counts())
x2=pd.DataFrame(df_netflix['release_year'].value_counts())
x3=pd.DataFrame(df_prime['release_year'].value_counts())
fnl_df=pd.concat([x1,x2,x3],axis=1)
fnl_df.reset_index(inplace=True)
fnl_df['total_count'] = fnl_df.sum(axis=1)
fnl_df=fnl_df.rename(columns={"index":"Year"})
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x="Year",y="total_count",data=fnl_df)
plt.show()
```
